### Convert data in proper format

```python
date_col = ['issue_date','last_credit_pull_date','next_payment_date']

for col in date_col :
    df[col]=pd.to_datetime(df[col],errors='coerce',dayfirst=True)
```

### Total applications

```python
total_applications=df.shape[0]
total_applications
```

```python
state_wise=df['address_state'].value_counts().reset_index()
state_wise
```

```python
top_5_states = df['address_state'].value_counts().head(5)

# Plot bar chart
plt.figure(figsize=(8,5))

top_5_states.plot(kind='bar', color='skyblue')
plt.title('Top 5 States by Loan Applications')
plt.xlabel('State')
plt.ylabel('No. of applications')
plt.xticks(rotation=45)
plt.show()
```

### Loan status distribution

```python
loan_status = df['loan_status'].value_counts(normalize = True) * 100
loan_status
```

```python
loan_status = df['loan_status'].value_counts(normalize = True) * 100

# Plot the pie chart
plt.figure(figsize=(5,5))
plt.pie(loan_status,
        labels=loan_status.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['skyblue', 'lightgreen', 'salmon', 'orange'])
plt.title('Loan Status Distribution')
plt.show()
```

### Grade wise Analysis

```python
Grade_wise = df.groupby('grade')['loan_status'].value_counts(normalize = True).unstack().fillna(0) * 100
Grade_wise
```

### Purpose wise loan applications

```python
purpose_wise = df['purpose'].value_counts(normalize = True) * 100
purpose_wise
```

```python
plt.figure(figsize=(10,6))
purpose_wise.plot(kind='bar', color='lightgreen', legend=False)

plt.title('Loan Purpose-wise Application', fontsize=14)
plt.xlabel('Loan Purpose', fontsize=12)
plt.ylabel('Number of Loans', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Verification status wise applications

```python
application_verified = df['verification_status'].value_counts(normalize = True) * 100
application_verified
```

```python
plt.figure(figsize=(3,3))

plt.pie(application_verified,
        labels=application_verified.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['skyblue', 'lightgreen', 'orange'])
plt.title('Loan Applications by Verification Status')

plt.show()
```

### Home Ownership Analysis

```python
home_ownership_dist = df['home_ownership'].value_counts(normalize = True) * 100
home_ownership_dist
```

### Profit or Loss per loan

```python
df['profit_or_loss'] = df['total_payment'] - df['loan_amount']

df[['id','loan_amount','total_payment','profit_or_loss','term']].sort_values(by='profit_or_loss',ascending = False).head(10) 
```
